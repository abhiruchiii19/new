Comparative analysis of ensembles techniques in machine learning ADABoost, Gradient Boost, XGboost, CATBoost

# **Introduction:**

Ensemble learning is a powerful approach in machine learning where multiple models are combined to produce a stronger and more robust predictive model than the individual models. Each model in the ensemble contributes to the final prediction, and the combination of diverse models often leads to improved generalization and performance.

In this practical, we focus on using ensemble methods, specifically AdaBoost, Gradient Boosting, XGBoost, and CatBoost, to classify iris flowers based on their features. AdaBoost, short for Adaptive Boosting, is an iterative ensemble technique that combines the predictions of weak learners (typically decision trees) to create a strong classifier. Gradient Boosting, XGBoost, and CatBoost are also ensemble methods that build models sequentially, with each new model correcting errors made by the previous ones.

```python
# Load libraries
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

1. **AdaBoost:**
   - AdaBoost combines multiple weak learners to create a strong classifier.
   - Weak learners are typically decision trees with a depth of 1 (stumps).
   - Each weak learner focuses on instances that were misclassified by the previous ones.


```python
abc = AdaBoostClassifier(n_estimators=50,
                         learning_rate=1)
model = abc.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn import metrics
svc=SVC(probability=True, kernel='linear')
abc =AdaBoostClassifier(n_estimators=50, base_estimator=svc,learning_rate=1)
model = abc.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Set the style for better visualization
sns.set(style="whitegrid")

# Create individual plots for each technique
plt.figure(figsize=(15, 8))

# Plot for AdaBoost with SVC
plt.subplot(2, 3, 1)
sns.heatmap(metrics.confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('AdaBoost')
```


2. **Gradient Boosting:**
   - Gradient Boosting builds trees sequentially, where each tree corrects the errors of the previous ones.
   - It uses a gradient descent optimization algorithm to minimize the loss function.
   - Popular implementations include scikit-learn's GradientBoostingClassifier.


```python
gb = GradientBoostingClassifier(n_estimators=50, learning_rate=1, max_depth=1, random_state=42)
model_gb = gb.fit(X_train, y_train)
y_pred_gb = model_gb.predict(X_test)
print("Gradient Boosting Accuracy:", metrics.accuracy_score(y_test, y_pred_gb))
```

```python

plt.subplot(2, 3, 3)
sns.heatmap(metrics.confusion_matrix(y_test, y_pred_gb), annot=True, fmt='d', cmap='Reds', cbar=False)
plt.title('Gradient Boosting')

```


3. **XGBoost (Extreme Gradient Boosting):**
   - XGBoost is an efficient and scalable implementation of gradient boosting.
   - It incorporates regularization terms to control overfitting and is designed for speed and performance.
   - It has become a popular choice in machine learning competitions.



```python
xgb = XGBClassifier(n_estimators=50, learning_rate=0.1, max_depth=3, random_state=42)
model_xgb = xgb.fit(X_train, y_train)
y_pred_xgb = model_xgb.predict(X_test)
print("XGBoost Accuracy:", metrics.accuracy_score(y_test, y_pred_xgb))
```

```python

# Plot for XGBoost
plt.subplot(2, 3, 4)
sns.heatmap(metrics.confusion_matrix(y_test, y_pred_xgb), annot=True, fmt='d', cmap='Oranges', cbar=False)
plt.title('XGBoost')
```

4. **CatBoost:**
   - CatBoost is a gradient boosting library that is particularly effective with categorical features.
   - It handles categorical variables automatically and efficiently.
   - CatBoost is known for its out-of-the-box performance and speed.

```python
catboost = CatBoostClassifier(iterations=50, learning_rate=0.1, depth=7, random_seed=42)
model_catboost = catboost.fit(X_train, y_train)
y_pred_catboost = model_catboost.predict(X_test)
print("CatBoost Accuracy:", metrics.accuracy_score(y_test, y_pred_catboost))
```

```python
# Plot for CatBoost
plt.subplot(2, 3, 5)
sns.heatmap(metrics.confusion_matrix(y_test, y_pred_catboost), annot=True, fmt='d', cmap='Purples', cbar=False)
plt.title('CatBoost')
```
